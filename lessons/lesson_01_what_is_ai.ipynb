### Simple Definitions

**AI (Artificial Intelligence)**
- The big umbrella term
- Any computer program that acts "smart"
- Example: A chess program that can beat humans

**Machine Learning (ML)**
- A type of AI that learns from data
- Instead of programming rules, you show it examples
- Example: Show it 1000 cat photos, it learns to recognize cats

**Deep Learning**
- A type of Machine Learning using "neural networks"
- Inspired by how the human brain works
- Very good at complex patterns (images, speech, language)

**LLM (Large Language Model)**
- A type of Deep Learning focused on language
- "Large" because they're trained on massive amounts of text
- This is what powers ChatGPT, Claude, and Gemini!

## Part 3: How Do LLMs (like ChatGPT) Actually Work?

### The Simple Analogy: The World's Best Autocomplete

You know how your phone suggests the next word when you're typing a text?

```
You type: "I'm going to the..."
Phone suggests: "store" "gym" "doctor"
```

**LLMs work the same way, but incredibly more advanced!**

Instead of just predicting one word, they predict entire paragraphs, essays, and conversations.

### How They Learn

1. **Training Phase** (happens before you use it)
   - The AI reads BILLIONS of web pages, books, articles
   - It learns patterns: what words typically follow other words
   - It learns facts, grammar, reasoning patterns
   - This takes months and millions of dollars!

2. **Using Phase** (when you chat with it)
   - You type a message (called a "prompt")
   - The AI predicts what text should come next
   - It generates one word at a time, very fast
   - This is why responses appear to "stream" in

### A Visual Example

```
Your prompt: "The capital of France is"

AI's internal process:
├── What word is most likely to come next?
│   ├── "Paris" → 95% probability ✓
│   ├── "London" → 0.1% probability
│   ├── "pizza" → 0.0001% probability
│   └── ... millions of other options
│
└── Pick "Paris", add it, repeat for next word...

Final output: "The capital of France is Paris."
```

## Part 4: Key Concepts You Need to Know

Before we start coding with AI, let's learn some important terms:

### Vocabulary List

| Term | Simple Definition | Example |
|------|------------------|----------|
| **Prompt** | The message you send to the AI | "Write me a poem about dogs" |
| **Response** | What the AI sends back | The poem it writes |
| **Model** | The specific AI "brain" you're using | GPT-4, GPT-4o-mini, Claude |
| **Token** | A piece of text (roughly a word) | "Hello" = 1 token, "artificial" = 1 token |
| **API** | A way for programs to talk to AI | How your code connects to ChatGPT |
| **Context** | Information the AI remembers during a chat | Previous messages in conversation |

### Understanding Tokens

Tokens are how AI "sees" text. Roughly:
- 1 token ≈ 4 characters in English
- 100 tokens ≈ 75 words

Why does this matter?
- **Pricing**: You often pay per token
- **Limits**: Models have maximum token limits

```
"Hello, how are you?" = 6 tokens
"Supercalifragilisticexpialidocious" = 7 tokens (long word = more tokens)
```

## Part 6: Common Misconceptions

Let's clear up some myths about AI:

### Myth 1: "AI understands like humans do"
**Reality**: AI doesn't truly "understand" - it recognizes patterns. It can produce text that LOOKS like understanding, but it's sophisticated pattern matching.

### Myth 2: "AI is always right"
**Reality**: AI can be confidently WRONG. This is called "hallucination." Always verify important information!

### Myth 3: "AI will replace all jobs"
**Reality**: AI is a tool that makes people more productive. People who USE AI will replace people who don't.

### Myth 4: "AI is too hard to learn"
**Reality**: Basic AI coding is easier than you think! If you can write simple Python, you can use AI.

### Myth 5: "AI remembers everything"
**Reality**: Most AI has limited memory. Each conversation starts fresh (unless you specifically design for memory).
