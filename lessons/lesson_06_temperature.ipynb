## Part 1: What is Temperature?

### The Concept

Remember from Lesson 1 how AI predicts the next word? Temperature affects HOW it picks words:

```
Prompt: "The best color is ___"

AI calculates probabilities:
- "blue" → 25%
- "red" → 20%
- "green" → 15%
- "purple" → 10%
- ... and thousands more

Temperature affects the selection:

LOW TEMPERATURE (0.0-0.3):        HIGH TEMPERATURE (0.8-1.2):
Almost always picks "blue"         More random - might pick "purple"
Very consistent                    More varied and creative
Same prompt → same response        Same prompt → different responses
```

### The Temperature Scale

| Temperature | Behavior | Best For |
|-------------|----------|----------|
| 0.0 | Almost deterministic | Facts, code, math |
| 0.3 | Very consistent | Customer support, Q&A |
| 0.7 | Balanced (default) | General conversation |
| 1.0 | Creative | Storytelling, brainstorming |
| 1.2+ | Very random | Wild creativity (can be nonsensical) |


---

## Part 2: Setup


````python
# Setup - Run this first!

from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True)
client = OpenAI()

def chat(prompt, temperature=0.7):
    """Chat with adjustable temperature"""
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        temperature=temperature  # Here's the new parameter!
    )
    return response.choices[0].message.content

print("✅ Ready to explore temperature!")
````

---

## Part 3: See Temperature in Action

Let's ask the same question multiple times at different temperatures.


````python
# Low temperature (0.0) - Very consistent

prompt = "Complete this sentence with one word: The sky is ___"

print("LOW TEMPERATURE (0.0) - Running 3 times:")
print("-" * 40)
for i in range(3):
    response = chat(prompt, temperature=0.0)
    print(f"  Attempt {i+1}: {response}")

print("\n↑ Notice how consistent the responses are!")
````

````python
# High temperature (1.0) - More varied

print("HIGH TEMPERATURE (1.0) - Running 3 times:")
print("-" * 40)
for i in range(3):
    response = chat(prompt, temperature=1.0)
    print(f"  Attempt {i+1}: {response}")

print("\n↑ Notice more variety in responses!")
````

---

## Part 4: Practical Examples

Let's see when to use different temperatures.


````python
# Example 1: Factual Question - Use LOW temperature

factual_prompt = "What is the capital of Japan?"

print("FACTUAL QUESTION")
print("=" * 50)
print(f"Prompt: {factual_prompt}")
print(f"\nLow Temp (0.0): {chat(factual_prompt, temperature=0.0)}")
print(f"\nHigh Temp (1.0): {chat(factual_prompt, temperature=1.0)}")

print("\n💡 For facts, use LOW temperature - you want the right answer!")
````

````python
# Example 2: Creative Task - Use HIGH temperature

creative_prompt = "Invent a name for a magical pet shop. Just give the name."

print("CREATIVE TASK")
print("=" * 50)
print(f"Prompt: {creative_prompt}")
print("\nLow Temperature (0.0) - 3 attempts:")
for i in range(3):
    print(f"  {chat(creative_prompt, temperature=0.0)}")

print("\nHigh Temperature (1.0) - 3 attempts:")
for i in range(3):
    print(f"  {chat(creative_prompt, temperature=1.0)}")

print("\n💡 For creativity, use HIGH temperature - you want variety!")
````

## Part 5: Project - Creative vs Factual Comparison Tool

Let's build a tool that shows the difference between creative and factual responses!


````python
# PROJECT: Creative vs Factual Response Comparison

def compare_temperatures(prompt, low_temp=0.0, high_temp=1.0):
    """
    Compare AI responses at different temperatures.
    
    Args:
        prompt: The question or task
        low_temp: Low temperature value
        high_temp: High temperature value
    """
    print("🎚️ TEMPERATURE COMPARISON")
    print("=" * 60)
    print(f"Prompt: {prompt}")
    print("=" * 60)
    
    print(f"\n❄️ LOW TEMPERATURE ({low_temp}) - Consistent/Factual:")
    print("-" * 40)
    print(chat(prompt, temperature=low_temp))
    
    print(f"\n🔥 HIGH TEMPERATURE ({high_temp}) - Creative/Varied:")
    print("-" * 40)
    print(chat(prompt, temperature=high_temp))

print("✅ Comparison tool ready!")
````

````python
# Test with different prompts!

compare_temperatures("Write a one-line description of a sunset.")
````

````python
# Try a factual vs creative prompt

compare_temperatures("Explain what a variable is in programming in one sentence.")
````

````python
# Your turn - try your own prompt!
compare_temperatures("Describe a perfect pizza.")
````

## Summary: Key Takeaways

### Temperature Guide

| Use Case | Recommended Temp |
|----------|-----------------|
| Code generation | 0.0 - 0.2 |
| Factual Q&A | 0.0 - 0.3 |
| Customer support | 0.3 - 0.5 |
| General chat | 0.7 (default) |
| Creative writing | 0.8 - 1.0 |
| Brainstorming | 0.9 - 1.2 |

### Key Points

1. **Low temp (0-0.3)**: Consistent, predictable, factual
2. **Medium temp (0.5-0.7)**: Balanced
3. **High temp (0.8-1.2)**: Creative, varied, unpredictable
4. **Too high (>1.5)**: Can become nonsensical
