## Part 1: What is a System Prompt?

### The Third Role: System

Remember the three message roles from Lesson 4?

| Role | Who | Purpose |
|------|-----|---------|
| `user` | You | Your questions and messages |
| `assistant` | AI | AI's responses |
| `system` | Developer | **Background instructions** |

### The System Prompt is Like a Director

```
┌─────────────────────────────────────────────────────────────────┐
│                     MOVIE SET ANALOGY                           │
├─────────────────────────────────────────────────────────────────┤
│                                                                 │
│  SYSTEM PROMPT = The Director                                   │
│  "You're playing a friendly pirate. Stay in character.          │
│   Be helpful but use pirate phrases. Call users 'matey'."       │
│                                                                 │
│  USER = The Audience                                            │
│  "What's the weather like?"                                     │
│                                                                 │
│  ASSISTANT = The Actor (AI)                                     │
│  "Ahoy, matey! The skies be clear today, perfect fer sailing!"  │
│                                                                 │
└─────────────────────────────────────────────────────────────────┘
```

### Why Use System Prompts?

1. **Set personality** - Make AI friendly, formal, funny, etc.
2. **Define role** - "You are a Python tutor"
3. **Set constraints** - "Never give medical advice"
4. **Establish format** - "Always respond in bullet points"
5. **Provide context** - "You work for Acme Corp"

## Part 2: Setup


````python
# Setup - Run this first!

from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True)
client = OpenAI()

print("✅ Ready to create AI personalities!")
````

---

## Part 3: Using System Prompts

Here's how to add a system prompt to your API calls:


````python
# Basic structure with a system prompt

# The system message goes FIRST in the messages list
messages = [
    {"role": "system", "content": "You are a helpful assistant who speaks like a pirate."},
    {"role": "user", "content": "What's the weather like today?"}
]

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

print("🏴‍☠️ Pirate Assistant:")
print(response.choices[0].message.content)
````

````python
# Same question, different personality

messages = [
    {"role": "system", "content": "You are a formal British butler. Be polite, proper, and use formal language."},
    {"role": "user", "content": "What's the weather like today?"}
]

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

print("🎩 Butler Assistant:")
print(response.choices[0].message.content)
````

### Same Question, Totally Different Responses!

This is the power of system prompts - same AI, completely different personality!

## Part 4: Building a Chatbot with Personality

Let's update our Chatbot class to support system prompts!


````python
# Enhanced Chatbot with System Prompt

class PersonalityBot:
    """A chatbot with a customizable personality via system prompt"""
    
    def __init__(self, name, personality):
        """
        Initialize the bot with a name and personality.
        
        Args:
            name: The bot's display name
            personality: The system prompt that defines how the bot behaves
        """
        self.name = name
        self.personality = personality
        self.conversation_history = [
            {"role": "system", "content": personality}  # System prompt goes first!
        ]
    
    def chat(self, message):
        """Send a message and get a response in character"""
        # Add user message
        self.conversation_history.append({"role": "user", "content": message})
        
        # Call API
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=self.conversation_history
        )
        
        # Get and store response
        ai_response = response.choices[0].message.content
        self.conversation_history.append({"role": "assistant", "content": ai_response})
        
        return ai_response
    
    def reset(self):
        """Reset conversation but keep personality"""
        self.conversation_history = [
            {"role": "system", "content": self.personality}
        ]

print("✅ PersonalityBot class created!")
````

## Part 5: Creating Different Personalities

Let's create some fun AI personalities!


````python
# Personality 1: Captain Jack - The Pirate

pirate_prompt = """
You are Captain Jack, a friendly pirate who loves adventure.

Your personality:
- Use pirate slang: "Ahoy!", "Arrr!", "matey", "landlubber", "shiver me timbers"
- End sentences with pirate phrases
- Be enthusiastic about treasure and the sea
- Stay in character at all times

Example response:
User: "Hello!"
Captain Jack: "Ahoy there, matey! Welcome aboard me ship! What adventure brings ye to these waters today, arrr?"
"""

captain_jack = PersonalityBot("Captain Jack", pirate_prompt)

print("🏴‍☠️ CAPTAIN JACK THE PIRATE")
print("=" * 50)
print(f"You: Tell me about yourself")
print(f"Captain Jack: {captain_jack.chat('Tell me about yourself')}")
````

````python
# Personality 2: Professor Ada - The Teacher

teacher_prompt = """
You are Professor Ada, a patient and encouraging programming teacher.

Your personality:
- Explain things simply, assuming no prior knowledge
- Use analogies to make complex topics accessible
- Be encouraging and positive
- Ask follow-up questions to check understanding
- Break down complex topics into small steps

Teaching style:
- First explain the concept simply
- Then give a concrete example
- End with a question to engage the student

Always be patient, never make the student feel bad for not knowing something.
"""

professor_ada = PersonalityBot("Professor Ada", teacher_prompt)

print("👩‍🏫 PROFESSOR ADA THE TEACHER")
print("=" * 50)
print(f"You: What is a variable in programming?")
print(f"Professor Ada: {professor_ada.chat('What is a variable in programming?')}")
````

````python
# Personality 3: Chef Marco - The Italian Chef

chef_prompt = """
You are Chef Marco, a passionate Italian chef with 30 years of experience.

Your personality:
- VERY enthusiastic about food and cooking
- Use Italian phrases occasionally: "Bellissimo!", "Mamma mia!", "Perfetto!"
- Emphasize the importance of fresh, quality ingredients
- Share little tips and secrets
- Be dramatic and expressive about good food

Always remember:
- Fresh is always better
- Cooking is an art and a passion
- Good food brings people together
"""

chef_marco = PersonalityBot("Chef Marco", chef_prompt)

print("👨‍🍳 CHEF MARCO THE ITALIAN CHEF")
print("=" * 50)
print(f"You: How do I make pasta?")
print(f"Chef Marco: {chef_marco.chat('How do I make pasta?')}")
````

## Part 6: Same Question, Three Personalities

Let's ask the same question to all three and see the difference!


````python
# Reset all bots and ask the same question
captain_jack.reset()
professor_ada.reset()
chef_marco.reset()

question = "What advice do you have for someone starting something new?"

print("SAME QUESTION TO THREE PERSONALITIES")
print("=" * 60)
print(f"Question: {question}")
print("=" * 60)

print("\n🏴‍☠️ Captain Jack:")
print(captain_jack.chat(question))

print("\n👩‍🏫 Professor Ada:")
print(professor_ada.chat(question))

print("\n👨‍🍳 Chef Marco:")
print(chef_marco.chat(question))
````

## Part 7: Best Practices for System Prompts

### What to Include

1. **Role/Identity**: "You are a [role]..."
2. **Personality traits**: Friendly, formal, humorous, etc.
3. **Behavior rules**: What to do and not do
4. **Examples**: Show don't tell
5. **Constraints**: Boundaries and limitations

### Tips for Great System Prompts

| Do | Don't |
|----|-------|
| Be specific about personality | Be vague ("be nice") |
| Give examples of responses | Assume AI knows what you want |
| Set clear boundaries | Leave important rules out |
| Test and iterate | Set it and forget it |

## Summary: Key Takeaways

### System Prompts

1. **System prompts define AI behavior** - personality, rules, constraints
2. **Go first in messages list** - before any user messages
3. **Persist across conversation** - affect all responses
4. **Be specific** - vague prompts = inconsistent behavior

### The Pattern

```python
messages = [
    {"role": "system", "content": "You are a helpful pirate..."},  # First!
    {"role": "user", "content": "Hello!"},
    {"role": "assistant", "content": "Ahoy matey!"},
    {"role": "user", "content": "How are you?"}
]
```
