## Part 2: Setup

Let's set up our connection first (same as Lesson 2).


````python
# Setup - Run this cell first!

from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True)
client = OpenAI()

def chat(prompt):
    """Simple function to chat with AI"""
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content

print("✅ Ready to learn about prompts!")
````

## Part 3: Good vs Bad Prompts - Live Examples

Let's see the difference in action!


````python
# Example 1: Vague vs Specific

print("=" * 60)
print("EXAMPLE 1: Vague vs Specific")
print("=" * 60)

# BAD: Vague prompt
bad_prompt = "Write about dogs"
print("\n❌ BAD PROMPT:", bad_prompt)
print("\nAI Response:")
print(chat(bad_prompt)[:300] + "...")  # Truncated for readability

print("\n" + "-" * 60)

# GOOD: Specific prompt
good_prompt = "Write a 2-sentence description of why golden retrievers make great family pets."
print("\n✅ GOOD PROMPT:", good_prompt)
print("\nAI Response:")
print(chat(good_prompt))
````

````python
# Example 2: No Context vs With Context

print("=" * 60)
print("EXAMPLE 2: No Context vs With Context")
print("=" * 60)

# BAD: No context
bad_prompt = "What should I do?"
print("\n❌ BAD PROMPT:", bad_prompt)
print("\nAI Response:")
print(chat(bad_prompt)[:300] + "...")

print("\n" + "-" * 60)

# GOOD: With context
good_prompt = """I'm a beginner programmer learning Python. I just finished learning 
about variables and loops. What topic should I learn next?"""
print("\n✅ GOOD PROMPT:", good_prompt)
print("\nAI Response:")
print(chat(good_prompt))
````

````python
# Example 3: Unclear Format vs Clear Format

print("=" * 60)
print("EXAMPLE 3: Unclear Format vs Clear Format")
print("=" * 60)

# BAD: No format specified
bad_prompt = "Give me some healthy breakfast ideas"
print("\n❌ BAD PROMPT:", bad_prompt)
print("\nAI Response:")
print(chat(bad_prompt)[:400] + "...")

print("\n" + "-" * 60)

# GOOD: Clear format specified
good_prompt = """Give me 3 healthy breakfast ideas. 
Format each as:
- Name: [breakfast name]
- Time to make: [X minutes]
- Main ingredient: [ingredient]"""
print("\n✅ GOOD PROMPT:", good_prompt)
print("\nAI Response:")
print(chat(good_prompt))
````

## Part 4: The CRISP Framework for Great Prompts

Here's an easy framework to remember:

### C.R.I.S.P.

| Letter | Meaning | Example |
|--------|---------|---------|
| **C** | Context | "I'm a teacher preparing a lesson for 5th graders..." |
| **R** | Role (optional) | "Act as a nutrition expert..." |
| **I** | Instructions | "Write a poem about..." |
| **S** | Specifics | "...in exactly 4 lines, using rhymes" |
| **P** | Parameters | "Keep it family-friendly, no complex words" |

### Example Using CRISP

```
Context:    "I run a small bakery in Seattle."
Role:       "Act as a marketing expert."
Instruction:"Write a tagline for my business."
Specifics:  "Maximum 6 words, should mention freshness."
Parameters: "Avoid clichés like 'best in town'."
```


````python
# Let's try the CRISP framework!

crisp_prompt = """
Context: I run a small bakery in Seattle that specializes in sourdough bread.
Role: Act as a marketing expert.
Task: Write a tagline for my business.
Specifics: Maximum 6 words, should emphasize freshness and craft.
Parameters: Avoid clichés like "best in town" or "since 19xx".

Give me 3 options.
"""

print("Using the CRISP Framework:")
print(crisp_prompt)
print("\n" + "=" * 40)
print("AI Response:")
print(chat(crisp_prompt))
````

## Part 5: Common Prompt Patterns

Here are some patterns that work really well:

### Pattern 1: Step-by-Step
Ask AI to think through something step by step.

### Pattern 2: Examples (Few-Shot)
Show AI examples of what you want.

### Pattern 3: Role-Playing
Give AI a persona to act as.

### Pattern 4: Constraints
Tell AI what NOT to do.

Let's see each in action!


````python
# Pattern 1: Step-by-Step Thinking

step_prompt = """
Solve this problem step by step:

A store has a 20% off sale. If a jacket originally costs $80, and you also have 
a coupon for $10 off (applied after the sale discount), what's the final price?

Think through each step before giving the final answer.
"""

print("PATTERN 1: Step-by-Step")
print("=" * 40)
print(chat(step_prompt))
````

````python
# Pattern 2: Few-Shot (Giving Examples)

fewshot_prompt = """
Convert these informal sentences to professional email language.

Examples:
Informal: "Hey, can u send me that file?"
Professional: "Could you please send me the document at your earliest convenience?"

Informal: "That meeting was a total waste of time"
Professional: "I believe the meeting could have been more productive with a clearer agenda."

Now convert this:
Informal: "I'm gonna be late cuz of traffic"
Professional:
"""

print("PATTERN 2: Few-Shot (Examples)")
print("=" * 40)
print(chat(fewshot_prompt))
````

````python
# Pattern 3: Role-Playing

role_prompt = """
You are a enthusiastic fitness coach named Coach Mike. You're encouraging, 
use lots of energy in your responses, and always try to motivate people.

A client says: "I'm feeling lazy and don't want to work out today."

Respond as Coach Mike would.
"""

print("PATTERN 3: Role-Playing")
print("=" * 40)
print(chat(role_prompt))
````

````python
# Pattern 4: Constraints (What NOT to do)

constraint_prompt = """
Write a short advertisement for a new smartphone.

CONSTRAINTS:
- Do NOT use the words "revolutionary", "game-changing", or "cutting-edge"
- Do NOT use exclamation marks
- Do NOT mention competitors
- Keep it under 50 words
"""

print("PATTERN 4: Constraints")
print("=" * 40)
print(chat(constraint_prompt))
````

## Part 6: Project - Build a Story Generator!

Now let's build something cool - a story generator that creates custom short stories based on your inputs!


````python
# PROJECT: The Story Generator!

def generate_story(genre, main_character, setting, conflict):
    """
    Generate a custom short story based on user inputs.
    
    Args:
        genre: The type of story (adventure, mystery, comedy, etc.)
        main_character: Description of the hero
        setting: Where/when the story takes place
        conflict: The problem or challenge to overcome
    
    Returns:
        A short story as a string
    """
    
    prompt = f"""
    Write a short story (about 150 words) with these elements:
    
    Genre: {genre}
    Main Character: {main_character}
    Setting: {setting}
    Conflict: {conflict}
    
    Requirements:
    - Start with an engaging opening line
    - Include some dialogue
    - End with a satisfying conclusion
    - Keep the tone appropriate for the genre
    """
    
    return chat(prompt)

print("✅ Story Generator ready!")
````

````python
# Let's test our story generator!

print("📚 STORY TIME! 📚")
print("=" * 50)

story = generate_story(
    genre="adventure",
    main_character="a curious 10-year-old girl named Maya",
    setting="a magical library where books come to life",
    conflict="she must find a lost book before sunset or the library disappears forever"
)

print(story)
````

````python
# Try a different genre!

print("🔮 MYSTERY STORY 🔮")
print("=" * 50)

mystery_story = generate_story(
    genre="mystery",
    main_character="a retired detective named Frank",
    setting="a small coastal town in Maine",
    conflict="someone has been leaving cryptic messages at the lighthouse"
)

print(mystery_story)
````

## Summary: Key Takeaways

### What Makes a Good Prompt?

1. **Be Specific** - Vague prompts = vague answers
2. **Provide Context** - Tell AI what it needs to know
3. **Specify Format** - Tell AI HOW to respond
4. **Use Constraints** - Tell AI what to avoid
5. **Give Examples** - Show AI what you want (few-shot)

### The CRISP Framework
- **C**ontext - Background information
- **R**ole - Who should AI act as?
- **I**nstructions - What should AI do?
- **S**pecifics - Details and format
- **P**arameters - Limits and constraints
